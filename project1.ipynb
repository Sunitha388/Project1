***********************DATA CLEANING PROCESS***********

```python


import pandas as pd   
import re

# Combine genre files with genre label
file_paths = [
    "imdb_movies_Action_2024.csv",
    "imdb_movies_Adventure_2024.csv",
    "imdb_movies_Animation_2024.csv",
    "imdb_movies_fantasy_2024.csv",
    "imdb_movies_family_2024.csv"
]

# Read and combine all genre CSVs
all_dfs = [pd.read_csv(file) for file in file_paths]
df = pd.concat(all_dfs, ignore_index=True)
print("Combined genres successfully.")

# Check column names
print("Available columns:", df.columns)

# Ensure 'Duration' exists
if 'Duration' not in df.columns:
    raise KeyError("'Duration' column not found in the input data. Please check the input files.")

# 1️. Remove serial numbers from the title/regex used to replace string regular expression 
df['Title'] = df['Title'].astype(str).str.replace(r'^\d+\.\s+', '', regex=True)

# 2️. Convert duration to minutes
def convert_duration(duration):
    match = re.match(r'(?:(\d+)h\s*)?(?:(\d+)m)?', str(duration))
    if match:
        hours = int(match.group(1)) * 60 if match.group(1) else 0
        minutes = int(match.group(2)) if match.group(2) else 0
        return hours + minutes  #  output is an integer FORMAT or not check
    return None  # Return None for invalid/missing durations

#  conversion for 'Duration' column
df['Duration(in_mins)'] = df['Duration'].astype(str).apply(convert_duration).astype(float)  # Use float for NaN handling

# Drop original 'Duration' column (optional)
df = df.drop(columns=['Duration'])


#VOTES CONVERTING
def convert_votes(vote_str):
    vote_str = vote_str.lower().strip()

    # Remove parentheses like (47k)  OR EXTRA SPACE IRUNDA
    vote_str = vote_str.replace('(', '').replace(')', '').strip()


    if vote_str.endswith('k'):
        return int(float(vote_str[:-1]) * 1000)   #rmoves last strng k(:-1)
    elif vote_str.endswith('m'):
        return int(float(vote_str[:-1]) * 1000000)
    else:
        try:
            return int(vote_str.replace(',', ''))
        except ValueError:
            return 0

df['Votes'] = df['Votes'].astype(str).apply(convert_votes)


# Remove rows with invalid/missing data
df = df[df["Duration(in_mins)"].notna()]  # 'Duration(in_mins)' is not NaN
df = df[df["Duration(in_mins)"] > 0]      # Remove rows with duration <= 0
df = df.dropna(subset=["Rating", "Votes"])  # Drop rows with missing 'Rating' or 'Votes'and checking

# Save the cleaned DataFrame to a CSV file
df.to_csv("file_imdb.csv", index=False)
print("Data is cleaned and saved in file_imdb.csv.")

#CHECKING 
print(df.head())
#print(df.columns)
```
